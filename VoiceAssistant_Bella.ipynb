# Imports

```python
import speech_recognition as sr
from gtts import gTTS
import playsound
import os
import pyaudio
import requests
import yfinance as yf
import wolframalpha
import time
import wikipedia
import datetime
```

# Voice Assistant Coding Section

```python
#Cryptocurrency API
crypto_api='https://api.coingecko.com/api/v3/simple/price?ids=bitcoin%2Clitecoin%2Csolana&vs_currencies=usd'


#Wolfram Alpha API Key
wolfram_api='7AE9V5-XJ6LTTTGRJ'

#Chuck Norris API
chuck_api='https://api.chucknorris.io/jokes/random'

#News API Key
news_api='bd209d5f0ddc47f6ae273e05e3efcef4'

#Weather API Key
weather_api='e5c7b526e9604b2ebdfe0cb94a3b1a2a'

#Distance API Key
distance_api='p5FGfGkZLtuwy5JNV3mG8tmF6vAczKqs'

def country_capital(text):
    client=wolframalpha.Client(wolfram_api)
    result=client.query(text)
    answer=next(result.results).text
    split=answer.split()
    capital='The capital of '+split[-1]+' is '+split[0]
    print(capital)
    bella_talk(capital)
    
def calculator(text):
    client=wolframalpha.Client(wolfram_api)
    result=client.query(text)
    answer=next(result.results).text
    print('The answer is '+answer)
    bella_talk('The answer is '+answer)
    
def president(text):
    client=wolframalpha.Client(wolfram_api)
    result=client.query(text)
    answer=next(result.results).text
    print('The president is '+answer)
    bella_talk('The president is '+answer)
    
def chuck_norris():
    cn_data=requests.get(chuck_api)
    cn_json=cn_data.json()
    joke=cn_json['value']
    print(joke)
    bella_talk(joke) 
    
def get_news():
    news_url='https://newsapi.org/v2/top-headlines?country=us&apiKey='+news_api
    news=requests.get(news_url).json()
    articles=news['articles']
    
    headlines=[]
    for i in articles:
        headlines.append(i['title'])
    
    for i in range(3):
        print(headlines[i])
        bella_talk(headlines[i])
        
def get_weather():
    bella_talk('No problem, I will look it up for you. What city are you interested in?')
    input1=bella_listen()
    weather_url='https://api.weatherbit.io/v2.0/current?city='+input1+'&key='+weather_api
    weather=requests.get(weather_url).json()
    temp=weather['data'][0]['temp']
    desc=weather['data'][0]['weather']['description']
    result='The temperature in '+input1+' is '+str(temp)+' degrees and you can see '+desc
    bella_talk(result)
    
def get_distance():
    bella_talk('Sure, let me know the starting point')
    from_city=bella_listen()
    time.sleep(1)
    bella_talk('Alright, let me know the final destination now.')
    to_city=bella_listen()
    bella_talk('Give me one moment, I will use my smart brain to calculate the distance for you.')
    dist_url='http://www.mapquestapi.com/directions/v2/route?key='+distance_api+'&from='+from_city+'&to='+to_city+'&unit=k'
    dist_request=requests.get(dist_url).json()
    distance=round(dist_request['route']['distance'],2)
    print(distance)
    dist_result='The distance between '+from_city+' and '+to_city+' is '+str(distance)+' kilometers '
    bella_talk(dist_result)
    
def wikipedia_info():
    bella_talk('I am happy to help. Let me know what I should search for you on Wikipedia.')
    wiki=bella_listen()
    wiki_result=wikipedia.summary(wiki,sentences=1)
    bella_talk(wiki_result)
    
def time_now():
    today_date=datetime.datetime.now()
    hour=today_date.strftime("%I")
    minute=today_date.strftime("%M")
    meridien=today_date.strftime("%p")
    time_now='It is '+hour+':'+minute+' '+meridien
    print(time_now)
    bella_talk(time_now)
    
def weekday_now():
    weekday_today=datetime.datetime.now().strftime("%A")
    print(weekday_today)
    bella_talk(weekday_today)


#convert speech to text so we can use the text further
def bella_listen():
    #create recognizer
    r=sr.Recognizer()
    #specifying source i.e. what we speak into our microphone
    with sr.Microphone() as source:
        #use the listen function so the recognizer can catch the source i.e. our mic
        audio=r.listen(source)
        text=''
        
        try:
            text=r.recognize_google(audio) #using google api to convert speech to text
            
        except sr.RequestError as re:
            print(re)
            
        except sr.UnknownValueError as uve:
            print(uve)
            
        except sr.WaitTimeoutError as wte:
            print(wte)

    text=text.lower()
    return text

#convert text to speech
def bella_talk(text):
    #create audio data
    file_name='audio_data.mp3'
    #convert text to speech
    tts=gTTS(text=text, lang='en')
    #save file
    tts.save(file_name)
    #play the audio file
    playsound.playsound(file_name)
    #remove file
    os.remove(file_name)
    
    
#reply function that will give back a reply based on the input text
def bella_reply(text):
    #smalltalk
    if 'what' in text and 'my name' in text:
        bella_talk('Your name is the most beautiful name I have ever heard.')
    elif 'what' in text and 'your name' in text:
        bella_talk('My name is Bella and I am your personal assistant.')
    elif 'why' in text and 'exist' in text:
        bella_talk('I was created to work for you. I dont need a break and I will never ask for days off.')
    elif 'when' in text and 'sleep' in text:
        bella_talk('I never sleep. I was created to support you 24 hours.')
    elif 'you' in text and 'stupid' in text:
        bella_talk('No i am not stupid. My grandmother told me there are no stupid persons out there.'+'I try to give my best everyday and learn continuously')
    elif 'favorite' in text or 'favourite' in text and 'movie' in text:
        bella_talk('My favourite movie is Titanic. I watch it with my friends all the time.')
        
        
    elif 'bitcoin' in text:
        response=requests.get(crypto_api)
        crypto_json=response.json()
        bella_talk('The current price for a Bitcoin is '+ str(crypto_json['bitcoin']['usd'])+' US Dollars')
    elif 'solana' in text:
        response=requests.get(crypto_api)
        crypto_json=response.json()
        bella_talk('The current price for a Solana is '+ str(crypto_json['solana']['usd'])+' US Dollars') 
    elif 'litecoin' in text:
        response=requests.get(crypto_api)
        crypto_json=response.json()
        bella_talk('The current price for a Litecoin is '+ str(crypto_json['litecoin']['usd'])+' US Dollars')
        
        
    elif 'apple' in text:
        apple= yf.Ticker('AAPL')
        bella_talk('At this moment, you can purchase one Apple share for '+str(apple.info['regularMarketPrice'])+' US Dollars')
    elif 'facebook' in text:
        fb= yf.Ticker('FB')
        bella_talk('At this moment, you can purchase one Facebook share for '+str(fb.info['regularMarketPrice'])+' US Dollars')
    elif 'tesla' in text:
        tesla= yf.Ticker('TSLA')
        bella_talk('At this moment, you can purchase one Tesla share for '+str(tesla.info['regularMarketPrice'])+' US Dollars')
        
    elif 'capital' in text and 'of' in text:
        country_capital(text)
    elif '+' in text or '-' in text or 'multiply' in text or 'multiplied' in text or 'divide' in text or 'root' in text:
        calculator(text)
    elif 'who' in text and 'president' in text:
        president(text)
        
    elif 'chuck norris' in text:
        chuck_norris()
    
    elif 'news' in text:
        bella_talk('Alright, let me tell you the top three headlines')
        get_news()
        
    elif 'weather' in text:
        get_weather()
        
    elif 'distance' in text:
        get_distance()
        
    elif 'wikipedia' in text:
        wikipedia_info()
        
    elif 'time' in text:
        time_now()
        
    elif 'weekday' in text:
        weekday_now()
        
    elif 'stop' in text:
        bella_talk('It was a pleasure to help you. I wish you a wonderful day.')
    else:
        bella_talk('Excuse me, I did not get that. Can you please repeat it?')
        
    
```

# Voice Assistant Execution

```python
def execute_bella():
    #personalize name
    bella_talk('Hi, I am here to support you. Can you please tell me your name?')
    name=bella_listen()
    bella_talk('Hi '+name+" What can i do for you?")
    
    while True:
        listen=bella_listen()
        print(listen)
        bella_reply(listen)
        
        if 'stop' in listen:
            break
#call the function
execute_bella()
```
